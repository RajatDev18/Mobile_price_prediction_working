```python
num_col = ['battery_power','clock_speed','fc','int_memory','m_dep','mobile_wt','px_height','px_width','ram','sc_h','sc_w','talk_time']
cat_col = ['blue','dual_sim','four_g','three_g','touch_screen','wifi','pc']
```

# Univariate Analysis

```python
for col in num_col:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
    sns.boxplot(x=df[col])
    plt.title(f"BoxPlot of {col}")
    plt.show()
```

```python
for col in cat_col:
    plt.figure(figsize=(10,4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count Plot of {col}")
    plt.ylabel("Count")
    plt.xlabel(col)
    plt.show()
```

```python
df.head()
```

# Bivariate Analysis

Numerical vs Target

```python
plt.figure(figsize=(20, 15))
for i, col in enumerate(num_col, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(x='price_range', y=col, data=df)
    plt.title(col)

plt.tight_layout()
plt.show()
```

Categorical vs Target

```python
for col in cat_col:
    pd.crosstab(df[col], df['price_range'], normalize='index').plot(kind='bar', stacked=True)
    plt.title(col)
    plt.show()
```
